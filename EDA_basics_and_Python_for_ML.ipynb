# EDA basics and python for Machine Learning

Basics of NumPy, Pandas, and data visualization with Matplotlib and Seaborn.

## NumPy Basics

```python
import numpy as np

# Creating arrays
a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]])

print("Array A:", a)
print("Array B Shape:", b.shape)
print("Dot Product:", np.dot(a[:2], b[:,0]))
```

## Pandas Basics

```python
import pandas as pd

# Load the Iris dataset from Seaborn's GitHub repository
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
print(df.head())

# Group by species and calculate the mean
print("Grouped mean by species:\n", df.groupby("species").mean())
```

## Data Visualization with Matplotlib & Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of sepal length
sns.histplot(df["sepal_length"], kde=True)
plt.title("Sepal Length Distribution")
plt.show()
```

## Visulization: Titanic Dataset Analysis

Load the Titanic dataset. Using `info()` and `describe()` funcs to  displays basic information about the data, including the data types and non-null counts.

> Add blockquote



```python
# Load Titanic dataset
titanic_df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Show basic info
print(titanic_df.info())
print('-------------------------------------------------------------------')
print(titanic_df.describe())

# Check null values
print("----Null Values----")
print(titanic_df.isnull().sum())
```

The code cell below visualizes the distribution of passenger ages using a histogram and also shows the survival rate by gender using a bar plot.

```python
# Visualize age distribution
sns.histplot(titanic_df['Age'].dropna(), kde=True)
plt.title("Age Distribution of Titanic Passengers")
plt.show()

# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title("Survival Rate by Gender")
plt.show()
```
