### Data Loading

The code below loads the data from the provided URL into a pandas DataFrame and displays the first few rows to inspect the data structure.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "http://bit.ly/w-data"
df = pd.read_csv(url)
print(df.head())
```

### Data Visualization

The scatter plot below visualizes the relationship between the number of hours studied and the scores obtained. Each point represents a student, with the x-axis showing the hours studied and the y-axis showing the scores.

```python
# Scatter plot
plt.scatter(df['Hours'], df['Scores'], color='blue')
plt.title("Hours Studied vs Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

### Model Training

The code below trains a linear regression model using the training data. The data is split into training and testing sets to evaluate the model's performance.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = df[['Hours']]
y = df['Scores']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)
```

### Prediction

The code below makes predictions on the test data using the trained linear regression model.

```python
y_pred = model.predict(X_test)

# Compare predictions
pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
```

### Regression Line

The code below visualizes the regression line, which represents the linear relationship between hours studied and scores, overlaid on the scatter plot of the data.

```python
line = model.coef_ * X + model.intercept_

plt.scatter(X, y, color='blue')
plt.plot(X, line, color='red')
plt.title("Regression Line - Hours vs Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

### Model Evaluation

The code below evaluates the performance of the trained linear regression model using metrics such as Mean Absolute Error, Mean Squared Error, and R-squared Score.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("Mean absolute error :", mean_absolute_error(y_test, y_pred))
print(" Mean squared error :", mean_squared_error(y_test, y_pred))
print("          R^2 Score :", r2_score(y_test, y_pred))
```

### Predicting with the Model

The code below demonstrates how to use the trained model to predict the score for a specific number of hours studied.

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

hours = 9.25
predicted_score = model.predict([[hours]])
print(f"Predicted score for {hours} hours of study: {predicted_score[0]:.2f}")
```
